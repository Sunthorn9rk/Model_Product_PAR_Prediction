# Load in Data

```python
import pandas as pd

Data_Par_qty_train_1 = pd.read_csv('/kaggle/input/dataset-find-par-qty-coop/train_fulldata.csv')
Data_Par_qty_train_1
```

# Data Preparation

```python
#To remove the NA values
print("จำนวนข้อมูลทั้งหมดก่อนทำการลบค่าที่ขาดหาย:", len(Data_Par_qty_train_1))
Data_Par_qty_train_1 = Data_Par_qty_train_1.dropna()
print("จำนวนข้อมูลทั้งหมดหลังทำการลบค่าที่ขาดหาย:", len(Data_Par_qty_train_1))
```

```python
# checking the number of missing values in the daataset
Data_Par_qty_train_1.isnull().sum()
```

```python
print("จำนวนข้อมูลทั้งหมดก่อนทำการลบค่าที่ซ้ำกัน :", len(Data_Par_qty_train_1))
# ลบแถวที่มีค่าข้อมูลที่ซ้ำกันออก
Data_Par_qty_train_1 = Data_Par_qty_train_1.drop_duplicates()

# ตรวจสอบจำนวนข้อมูลหลังจากลบแถวที่ซ้ำกัน
print("จำนวนข้อมูลทั้งหมดหลังทำการลบค่าที่ซ้ำกัน:", len(Data_Par_qty_train_1))
```

# Split Data

```python
Data_Par_qty_train_1
```

```python
from sklearn.model_selection import train_test_split
#Splitting dataset to train and test
X_1 = Data_Par_qty_train_1.drop(['Parqty_Cal','Parqty_Adj'], axis=1) # features 

y_1 = Data_Par_qty_train_1[['Parqty_Cal']] # target

# แบ่งข้อมูลเป็น train 70% และ temp set 30%
X_train_1, X_temp_1, y_train_1, y_temp_1 = train_test_split(X_1, y_1, test_size=0.3, random_state=42)
  
# แบ่ง temp set เป็น validation 15% และ test 15%
X_val_1, X_test_1, y_val_1, y_test_1 = train_test_split(X_temp_1, y_temp_1, test_size=0.5, random_state=42)
```

# Train Model Round1

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# ฝึกโมเดลเริ่มต้นบนชุดฝึกอบรม
Model_RF_0 = RandomForestRegressor(n_estimators=10, random_state=42)
Model_RF_0.fit(X_train_1, y_train_1.values.ravel())

# ปรับค่าพารามิเตอร์บนชุดการตรวจสอบค่า hyperparameter
Model_RF_1 = RandomForestRegressor(n_estimators=15, criterion='squared_error'
                                  , min_samples_split=2, random_state=42)
Model_RF_1.fit(X_val_1, y_val_1.values.ravel()) 

# ประเมินประสิทธิภาพบนชุดทดสอบ
y_pred_1 = Model_RF_1.predict(X_test_1)

# ประเมินประสิทธิภาพของโมเดล
mse_Train_1 = mean_squared_error(y_test_1, y_pred_1)
print(f'Mean Squared Error บนชุดทดสอบ: {mse_Train_1}')
```

```python
Data_Train_Eli_col
```

```python
#Splitting dataset to train and test
X_3 = Data_Train_Eli_col.drop(['Parqty_Cal','Parqty_Adj'], axis=1) # features 

y_3 = Data_Train_Eli_col[['Parqty_Cal']] # target

# แบ่งข้อมูลเป็น train 70% และ temp set 30%
X_train_3, X_temp_3, y_train_3, y_temp_3 = train_test_split(X_3, y_3, test_size=0.3, random_state=42)

# แบ่ง temp set เป็น validation 15% และ test 15%
X_val_3, X_test_3, y_val_3, y_test_3 = train_test_split(X_temp_3, y_temp_3, test_size=0.5, random_state=42)
```

```python
Model_RF_3 = RandomForestRegressor(n_estimators=200, min_samples_split=2, min_samples_leaf= 1
                                  , max_features=1.0, max_depth=None ,bootstrap= False
                                  , random_state=42)
Model_RF_3.fit(X_val_3, y_val_3.values.ravel())

# ประเมินประสิทธิภาพบนชุดทดสอบ
y_pred_3 = Model_RF_3.predict(X_test_3)

# ประเมินประสิทธิภาพของโมเดล
mse_Train_3 = mean_squared_error(y_test_3, y_pred_3)
print(f'Mean Squared Error บนชุดทดสอบ: {mse_Train_3}')
```

# Measuring model performance

```python
Parqty_Cal_Pred = y_pred_4.round()

Diff = Data_Par_qty_Test_4['Parqty_Cal'] - Parqty_Cal_Pred

count_correct = (Diff == 0).sum()

len_Data_All = len(Data_Par_qty_Test_4)

accuracy = count_correct / len_Data_All

print('MSE Before Turning Hyperparameter : {:0.2f}'.format(mse_Train_1))
print('MSE Before Turning Hyperparameter : {:0.2f}'.format(mse_Train_3))
print('Improvement of MSE : {:0.2f}'.format(mse_Train_1 - mse_Train_3))
print('Final Accuracy : {:0.2f}%.'.format(count_correct / len_Data_All))
```
